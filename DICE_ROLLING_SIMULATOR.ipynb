```python
class Dice_Roller:
    def random_number(self):
        import random
        dice = random.randrange(1,7)
        if dice == 6:
            print('''
                     ------
                    |*    *|
                    |*    *|
                    |*    *|
                     ------ ''')
            print("\n\t\twhoah...游뱔")
                  
        elif dice == 1:
            print('''
                     -------
                    |       |
                    |   *   |
                    |       |
                     -------''')
            print("\n\t\tBetter Luck...游땛")
           
        elif dice == 5:
              print('''
                     -------
                    |*     *|
                    |   *   |
                    |*     *|
                     -------''')
                
        elif dice == 4:
              print('''
                     -------
                    |*     *|
                    |       |
                    |*     *|
                     -------''')
        elif dice == 3:
              print('''
                     -------
                    |*      |
                    |   *   |
                    |      *|
                     -------''')
        elif dice == 2:
              print('''
                     -------
                    |*      |
                    |       |
                    |      *|
                     -------''')
                
    def roll_dice(self):
        print("\t\t 游쓇릝 DICE ROLLING SIMULATOR 游쓇릝 \n")
        choice =input("\nRoll Dice  ").lower()
        print("\nRolling the dice...........游때")
        self.random_number()
       
            
        while True:
            choice =input("Roll  again? yes or Quit: ").lower()
            if choice == 'yes':
                print("\nRolling the dice...........游때")
                self.random_number()
            elif choice == 'quit':
                print('\n\tGame End....游쓇릝')
                break
            else:
                break
                
#creating an object
object = Dice_Roller()
object.roll_dice()
```
